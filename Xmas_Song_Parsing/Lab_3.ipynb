Objective: I need to turn singular words into the pural form. Each word has there own plural form. So I should input the word and get back the plural form

```python
def pluralize_gift(gift):
  """
  Returns plural of a noun

  Parameters
  ----------
  gift: str
    A noun

  Return
  ------
  str
    Plural version
  """

  if (gift[-1]=='y'):
    gift = gift.replace("y", "ies")

  elif ("oo" in gift):
    gift = gift.replace("oo", "ee")
  else:
    gift = gift + "s"

  return gift

```

```python
def make_phrase(num, num_word, item, verb, adjective, location):
    """
  Returns the phrases of the song 12 Days of Christmas

  Parameters
  ----------
  num: int
    day number
  num_word: str
    word representation of the day number
  item: str
    gift item
  verb: str
    action associated with the gift item
  adjective: str
    descriptor of the gift item
  location: str
    area where the gift item is found

  Return
  ------
  str
    full phrase
  """
    maped_words = {
        "first": "one",
        "second": "two",
        "third": "three",
        "fourth": "four",
        "fifth": "five",
        "sixth": "six",
        "seventh": "seven",
        "eighth": "eight",
        "ninth": "nine",
        "tenth": "ten",
        "eleventh": "eleven",
        "twelfth": "twelve"
    }

    # Step 1: Replace NAs with blank strings using a truth check
    verb = verb if pd.notna(verb) else "" # Resource for NA values: https://pandas.pydata.org/docs/reference/api/pandas.notna.html
    adjective = adjective if pd.notna(adjective) else ""
    #  assigns the value of the variable "adjective" to itself if it is not a NaN value; otherwise, it assigns an empty string.
    location = location if pd.notna(location) else ""

    # Step 2: If the day number is larger than 1, the gift items need pluralized!
    if not (num == 1):
        item = pluralize_gift(item)
   # Step 3: Get the numerical representation of the day word
    num_word = maped_words.get(num_word)

    # Step 4: Construct the day part of the phrase
    if num == 1:
        # Check if the gift item starts with a vowel
        starts_with_vowel = item[0] in ['a', 'e', 'i', 'o', 'u']

        # Use "an" if it starts with a vowel, otherwise use "a"
        num_to_word = "an" if starts_with_vowel else "a"
    else:
        num_to_word = num_word

    # Step 5: Put all of the pieces together into one string and return!
    phrase = f"{num_to_word} {adjective} {item} {verb} {location}".strip().replace("  ", " ")  # ensures there is only one space between words
    return phrase


```

#### Adding the formating to the sing_day function

1. Let's try the formated fucntion for the 12 days of christmas

```python
def sing_day(dataset, num, phrase_col):
    """
    Returns the verse of the song "12 Days of Christmas" for the specified day

    Parameters
    ----------
    dataset : DataFrame
        The dataset containing the phrases for each day.
    num : int
        The day number.
    phrase_col : str
        The column name in the dataset that contains the phrases for each day.

    Return
    ------
    str
        The song verse for the specified day.
    """
    num_to_word = {
        1: "first",
        2: "second",
        3: "third",
        4: "fourth",
        5: "fifth",
        6: "sixth",
        7: "seventh",
        8: "eighth",
        9: "ninth",
        10: "tenth",
        11: "eleventh",
        12: "twelfth"
    }

    # Step 1: Setup the intro line
    num_word = num_to_word[num]
    intro = f"On the {num_word} day of Christmas, my true love gave to me:"

    # Step 2: Sing the gift phrases
    gifts_list = dataset[phrase_col].iloc[:num].tolist()[::-1]

    # Special case: The second last gift gets ", and" appended to it if there's more than one gift of the day
    if num > 1:
        gifts_list[-1] = "and " + gifts_list[-1]

    # Add commas to all items except the last one, which gets a period
    gifts_list[:-1] = [gift + "," if ", and" not in gift else gift for gift in gifts_list[:-1]]
    gifts_list[-1] = gifts_list[-1] + "."

    # Combine the intro and gifts for the final song
    song = intro + "\n" + "\n".join(gifts_list)

    return song

# Testing the function with the earlier provided xmas dataset and phrase column
print(sing_day(xmas, 7, "Full.Phrase"))
```

Let's create one more function thats going to sing the full song depending on what day it it

1. Let's create a new fucntion that will sing the entire song

```python
def sing_full_song(dataset, num, phrase_col):
    """
    Returns the full song "12 Days of Christmas" up to the specified day without using for loops.

    Parameters
    ----------
    dataset : DataFrame
        The dataset containing the phrases for each day.
    num : int
        The ending day number.
    phrase_col : str
        The column name in the dataset that contains the phrases for each day.

    Return
    ------
    str
        The song verses up to the specified day.
    """
    # Use map() to apply sing_day function over a range of days starting from 1
    songs = list(map(lambda day: sing_day(dataset, day, phrase_col), range(1, num + 1)))

    # Join the songs
    full_song = "\n\n".join(songs)

    # Fix for the "and" on the first day
    full_song = full_song.replace(", and", ".")

    return full_song

# Testing the function with the earlier provided xmas dataset and phrase column
print(sing_full_song(xmas, 12, "Full.Phrase"))
```
