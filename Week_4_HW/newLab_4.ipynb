### 1. Scrape the Location Counts

Starbucks

```python
import requests
response = requests.get("https://www.menuism.com/restaurant-locations/starbucks-coffee-39564")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find_all('li',{'class':""})
len(table) # there are 5 major regions that we need to scrape data from so we need to use the location-list-header
```

```python
import re
import pandas as pd

# initialize an empty list
rows = []

# iterate over all rows in the faculty table
for coffee_location in soup.find_all('li',{'class':""}):
    a_tag = coffee_location.find("a")
    if a_tag:
        text = a_tag.get_text(strip=True)

        # use regex to extract state name and number of locations

        match = re.match(r"(.+) Starbucks Coffee locations \((\d+)\)", text) # use \( to excape the parenteses
        # https://www.guru99.com/python-regular-expressions-complete-tutorial.html
        if match:
            state_name = match.group(1)
            num_locations = int(match.group(2))

            # append this data
            rows.append({
                "State Names": state_name,
                "Store Locations": num_locations
            })

# create a DataFrame
pd.DataFrame(rows)



```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
def scrape_store_locations(url):
    """
    This function takes a URL of a webpage containing a list of store locations and returns a dataframe with three columns:
    - State Names: Abbreviations of US states
    - Store Locations: Number of store locations in the state
    - Company Name: Name of the company

    Parameters:
    url (str): The URL of the webpage containing the list of store locations.

    Returns:
    pd.DataFrame: A dataframe containing the store locations data.
    """
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    rows = []
    for coffee_location in soup.find_all('li', {'class': ""}):
        a_tag = coffee_location.find("a")
        if a_tag:
            text = a_tag.get_text(strip=True)
            us_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
                         "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
                         "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
                         "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
                         "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
            match = re.match(r"(.+?) (.+) locations \((\d+)\)", text)
            if match:
                state_name = match.group(1)
                company_name = match.group(2)
                num_locations = int(match.group(3))
                state_abbreviation = stateabb(state_name)
                if state_abbreviation in us_states:
                    rows.append({
                        "State Names": state_abbreviation,
                        "Store Locations": num_locations,
                        "Company Name": company_name
                    })
    df = pd.DataFrame(rows)
    return df
```

```python
from plotnine import *
stores_by_chain = coffee.groupby('Company Name')['Store Locations'].sum().reset_index()
stores_by_chain
```

```python
# 1. Total Number of Stores by Chain
plot1 = (
    ggplot(stores_by_chain, aes(x='Company Name', y='Store Locations', fill='Company Name')) +
    geom_bar(stat='identity') +
    geom_text(aes(label='Store Locations'), nudge_y=0.5, color='black') +
    labs(title='Total Number of Stores by Chain', x='Company Name', y='Number of Stores') +
    theme_matplotlib() +
    theme(legend_position='none',
           axis_text_x=element_text(rotation=45, hjust=1))
)
plot1

```

```python
# 2. Number of Stores by Chain and Region
stores_by_chain_region = coffee.groupby(['Company Name', 'region'])['Store Locations'].sum().reset_index()

plot3 = (
    ggplot(stores_by_chain_region, aes(x='region', y='Store Locations', fill='Company Name')) +
    geom_bar(stat='identity', position='dodge') +
    labs(title='Number of Stores by Chain and Region', x='Region', y='Number of Stores') +
    theme_minimal() +
    theme(legend_position='right')
)
plot3
```

```python
grouped_coffee = coffee.groupby(['State Names', 'Company Name'])['Store Locations'].sum().reset_index()

# Create a bar plot showing the number of stores for each coffee chain in each state
plot_state = (ggplot(grouped_coffee, aes(x='State Names', y='Store Locations', fill='Company Name')) +
              geom_bar(stat='identity', position='stack') +
              labs(title='Number of Stores for Each Coffee Chain in Each State',
                   x='State',
                   y='Number of Stores') +
              theme_tufte()+
              theme(figure_size=(15, 6),
                    axis_text_x=element_text(angle=90)))
plot_state
```

**Interpretation:**
Starbucks and McDonald's clearly have a stronghold in the coffee market across the United States, as evidenced by their widespread presence in nearly all regions and states, and their substantial revenue figures. This dominance is a reflection of their expansive number of outlets nationwide.Interestingly, despite having a lower overall store count compared to Starbucks and McDonald's, Dunkin' Donuts emerges as the premier coffee distributor in the Northeast region, particularly in Massachusetts where it holds a significant market share

A nuanced comparison between McDonald's and Starbucks reveals some regional variations in their prevalence. Starbucks has carved out a slightly more dominant position in California, whereas McDonald's is the undisputed leader in Texas. This pattern extends to regional dominance, with Starbucks being the predominant coffee chain in the West, and McDonald's reigning supreme in the South.

The data paints a fascinating picture of the regional tastes and market forces that dictate the prominence of these coffee chains in different parts of the country. It is evident that while the total number of stores is a substantial factor in their market presence, regional and state-specific dynamics also play a pivotal role in shaping their market dominance.
